```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
#Replace the prefix ans suffix with empty values
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
```

```python
print(student_data_df)
```

```python
#combining two dataframes into one
school_data_complete_df = pd.merge(student_data_df,school_data_df, on =["school_name","school_name"])
school_data_complete_df.head()
```

```python
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
school_count = school_data_df["School ID"].count()
school_count
```

```python
school_count2 = len(school_data_complete_df["school_name"].unique())
school_count2
```

```python
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#Get all the student who passed in maths in new DataFrame
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
#Get all the student who passed in reading in new DataFrame
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
#Calculate number of student passed in Maths
passing_math_count = passing_math["student_name"].count()
passing_math_count
```

```python
#Calculate number of students passed in reading
passing_reading_count = passing_reading["student_name"].count()
passing_reading_count
```

```python
#Calculate the percentage of students passed in Maths
passing_math_percentage = passing_math_count/float(student_count)*100
passing_math_percentage
```

```python
#Calculate the percentage of students passed in Reading
passing_reading_percentage = passing_reading_count/float(student_count)*100
passing_reading_percentage
```

```python
#Calculate students who passed in Maths and Reading
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)]
passing_math_reading
```

```python
#Calculate overall passing count
passing_math_reading_count = passing_math_reading["student_name"].count()
passing_math_reading_count
```

```python
#Calculate overall passing percentage
passing_math_reading_percentage = passing_math_reading_count/float(student_count)*100
passing_math_reading_percentage
```

```python
#Adding the list of values to DataFrame
district_summary_df = pd.DataFrame([{"Total_Schools": school_count,
                                    "Total_students": student_count,
                                    "Total_Budget":total_budget,
                                    "Average_Reading_Score":average_reading_score,
                                    "Average_Math_Score":average_math_score,
                                    "%_Passing_Reading":passing_reading_percentage,
                                    "%_Passing_Math":passing_math_percentage,
                                    "%_Overall_Passing":passing_math_reading_percentage}])
district_summary_df
```

```python
district_summary_df["Total_students"] = district_summary_df["Total_students"].map("{:,}".format)
district_summary_df["Total_students"]
```

```python
district_summary_df["Total_Budget"] = district_summary_df["Total_Budget"].map("${:,.2f}".format)
district_summary_df["Total_Budget"]
```

```python
district_summary_df["Average_Reading_Score"] = district_summary_df["Average_Reading_Score"].map("{:.1f}".format)
district_summary_df["Average_Reading_Score"]
```

```python
district_summary_df["Average_Math_Score"] = district_summary_df["Average_Math_Score"].map("{:.1f}".format)
district_summary_df["Average_Math_Score"]
```

```python
district_summary_df["%_Passing_Reading"] = district_summary_df["%_Passing_Reading"].map("{:.0f}%".format)
district_summary_df["%_Passing_Reading"]
```

```python
district_summary_df["%_Passing_Math"] = district_summary_df["%_Passing_Math"].map("{:.0f}%".format)
district_summary_df["%_Passing_Math"]
```

```python
district_summary_df["%_Overall_Passing"] = district_summary_df["%_Overall_Passing"].map("{:.0f}%".format)
district_summary_df["%_Overall_Passing"]
```

```python
new_column_order = ["Total_Schools","Total_students", "Total_Budget", "Average_Math_Score", "Average_Reading_Score", "%_Passing_Math", "%_Passing_Reading", "%_Overall_Passing"]

district_summary_df = district_summary_df[new_column_order]
district_summary_df
```
